  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
nums =[1,2,20]

def missing_num(nums: list[int]) -> int:
    n = len(nums)
    if n== 0:
        return ValueError("The input list is empty")
    max_num = nums[0]
    for i in range(n-1):
        if nums[i+1] > max_num:
            max_num = nums[i+1]
    supplemental_num =[]
    for i in range(max_num):
        if i not in nums:
            supplemental_num.append(i)
    if len(supplemental_num) == 0:
        return -1
    return supplemental_num
```

### Reflection

For Assignment 1, I implemented move_zeros_to_end(nums) with a single forward scan using a while loop. My idea was to keep the operation in-place and preserve the original order of non-zero elements (stability). Whenever I encountered a zero at index i, I removed it with pop(i) and immediately pushed it to the end with append, while not advancing i (because a new element slid into position i). When I saw a non-zero, I incremented i. I also tracked n (the effective boundary for scanning) and decreased it after moving a zero, so the loop only processed each position until all non-zeros had been ‚Äúcompacted‚Äù to the front. This approach was easy to reason about and required O(1) extra space, but I learned that repeated middle pop(i) operations make the worst-case time O(n¬≤) (e.g., many zeros near the front). I validated behavior on edge cases‚Äîempty list, all zeros, no zeros, and mixed values‚Äîand confirmed that stability holds.
